```java
class Findprimenumber{   
	public static void main(String[] args) {
		int a=1,b=100;
		for(int i=a;i<=b;i++){
		    if(checkPrime(i)){
		        System.out.print(i+" " );
		    }
		}
	}
	public static boolean checkPrime(int num){
	    // 0, 1 and negative numbers are not prime
	    if(num<2){
	        return false;
	    }
	    else{
	        int x= num/2;
	        for(int i=2;i<x;i++){
	            if(num%i==0){
	                return false;
	            }
	        }
	    }
	    return true;
	}
}
```

### Program to check given number is prime number or not
**BruteForce Approach**

```java
class Checkprime{
    public static void main(String[] args){
        Scanner sc = new Scanner(System.in);
        System.out.println("Enter number to check prime or not : ")
        int n = sc.nextInt();
        boolean prime = true;
        for(int i=2;i<n;i++){
            if(n%i==0){
                prime = false;
                break;
            }
        }
        if(prime){
            System.out.println("This number is prime number");
        }
        else{
            System.out.println("This number is not prime number");
        }
    }
}

```

**Good Approach**
```java
class Checkprime2{
    public static void main(String[] args){
        Scanner sc = new Scanner(System.in);
        System.out.println("Enter number to check prime or not : ")
        int n = sc.nextInt();
        boolean prime = true;

        // using start loop 2 to square root of the given number (i*i) or Math.sqrt(n),
        
        for(int i=2;i*i<n;i++){
            if(n%i==0){
                prime = false;
                break;
            }
        }
        if(prime){
            System.out.println("This number is prime number");
        }
        else{
            System.out.println("This number is not prime number");
        }

    }
}

```

**Better Approach**
```java
class Checkprime3{

    public static boolean check(int n){
        //firstly, check with divisiblity with 2 means (also with 4,8,10,12,14 .....) and 3 means (also with 6,9,12,15,18,21,.......)
        if(n%2==0 || n%3==0){
            return false;
        }
        // secondly, check with rest numbers i.e (5,7,11,13,17,19,23 .........)
        
        for(int i=5;i<n;i+=6){
            if(n%i==0 || n%i+2 == 0){
                return false;
                break;
            }
        }
        return true;
    }

    public static void main(String[] args){
        Scanner sc = new Scanner(System.in);
        System.out.println("Enter number to check prime or not : ")
        int n = sc.nextInt();
        // boolean prime = true;
        if(check(n) == true){
            System.out.println("This number is prime number.")
        }
        else{
            System.out.println("This number is not prime number.")
        }

    }
}
```

### program to check the number is armstrong number or not.

**Logic**
$$
n=153\\

```java
class checkarmstrog{
    public static void main(String[] args){
        Scanner sc = new Scanner(System.in);
        System.out.println("Enter number to check armstrong or not : ")
        int n = sc.nextInt();
        int digit = 0;
        int original = n;
        while(n!=0){
            digit+=Math.pow(n%10,3);
            n/=10;
        }
        if(digit == original){
            System.out.println("This number is armstrong number.")
        }
        else{
            System.out.println("This number is not arnstrong number.")
        }
    }
}
```
### Program to check the number is palindrome or not.

```java
class Checkpalindrome{
    public static void main(String[] args){
        Scanner sc = new Scanner(System.in);
        System.out.println("Enter number to check palindrome or not : ")
        int n = sc.nextInt();
        int rev = 0;
        int original = n;
        while(n!=0){
            rev = rev*10 + n%10;
            n/=10;
        }
        if(rev == original){
            System.out.println("This number is palindrome number.")
        }
        else{
            System.out.println("This number is not plaindrome number.")
        }
    }
}
```

## Recursion
- *Recusion is define as a function which call itself.*
- *when call the function it store in stack memory.*
- *Base condition are used to safe from **stackoverflow Error**.*

![images](recursion.png)

#### How to solve the recusion problem.
- let given problem is **big problem**.
- And find out the **small problem** from the big problem.
- start the **recursion cycle**.
- Terminate with **base condition**.

**Example**
Problem => print the number 1 to 10.
1. let 1 to 10 is big problem.
2. if we print 1 to 2 i.e small problem.
    - it should be count + 1
3. use recusion function call.
4. Add base condition i.e if(count == 10){  return; }

```java
// Count the number the digit
class Recursion{
    public static void count(int num, int digit){
        //Base condition or termination case
        if(num == 0){
            System.out.println(digit);
            return;
            // return recently previous function called
        }
        //when we count one number, then number have decrerase
        num = num/10;
        // count the number by +1
        count(num , digit + 1);
    }
    public static void main(String[] args){
        int num = 12345;
        // calling the recursion function
        count(num,0);
    }
}
```

**Problem-3**  => Factorial using recursion

```java
class Factorial{
  public static int fact(int num){
    // Base Condition
    if(num == 0 || num == 1){
      return 1;
    }
    return num * fact(num-1);
  }
  public static void main(String[] args){
    System.out.println("factorial is :" + fact(5));
  }
}
```

```java
class CheckPrimeNumber{
  // static int i = 2;
  public static boolean prime(int num, int i){
    //Base condition
    // condition 1 : if number is 2 i.e i -> return true
    // condition 2 : if number is divisible by 2 -> return false
    // condition 3 : if square of i is larger than number -> return true
    if(num < i){
      return false;
    }
    if(num == i){
      return true;
    }
    if(num % 2 == 0){
      return false;
    }
    if(i*i > num){
      return true;
    }
    return prime(num, i+1);
  }
  public static void main(String[] args){
    System.out.println(prime(7,2));
  }
}
```
